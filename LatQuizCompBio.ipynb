```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
```

```python
# working with biological genBank Files
# read (hanya 1 data dalam 1 file fasta/gb)
readF = SeqIO.read('sequence.fasta', 'fasta')
print(readF)

readGB = SeqIO.read('sequence.gb', 'gb')
print(readGB)

# parse (ada > 2 data dalam 1 file fasta/gb)
for parseF in SeqIO.parse('sequence.fasta', 'fasta'):
  print(parseF)

for parseGB in SeqIO.parse('sequence.gb', 'gb'):
  print(parseGB)
```

```python
# sequence manipulation using python
# slicing
# variable[idxawal : idxakhir + 1 : lompatan]
seqX = readF.seq

# seqA adalah setengah bagian pertama dr seqX

seqA = seqX[ : round(len(seqX) / 2)]

# seqB adalah setengah bagian terakhir dr seqX

seqB = seqX[round(len(seqX) / 2):]

```

```python
from Bio.SeqUtils import GC
from Bio.SeqUtils import MeltingTemp as mt
from Bio.SeqUtils import molecular_weight as mw
from collections import Counter
from matplotlib import pyplot as plt
```

```python
# !Sequence analysis!
# determine mt

mtA = mt.Tm_Wallace(seqA)
mtB = mt.Tm_Wallace(seqB)

print('mtA:', mtA)
print('mtB:', mtB)

# determine GC / AT

gcA = GC(seqA)
gcB = GC(seqB)

atA = 100 - gcA
atB = 100 - gcB

print('gcA:', gcA)
print('gcB:', gcB)
print('atA:', atA)
print('atB:', atB)

# determine mw

mwA = mw(seqA)
mwB = mw(seqB)

print('mwA:', mwA)
print('mwB:', mwB)

# finding nucleotide base freq
freqA = Counter(seqA)
freqB = Counter(seqB)

print(freqA)
print(freqB)
```

```python
# DNA & mRNA protein sintesis
# transcription
mrnaA = seqA.transcribe()
mrnaB = seqB.transcribe()

print('mrnaA:', mrnaA)
print('mrnaB:', mrnaB)

# translation
proteinA = mrnaA.translate()
proteinB = mrnaB.translate()

print('proteinA:', proteinA)
print('proteinB:', proteinB)
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment as fa
```

```python
# find sequence alignments and similarities

# bikin function dulu buat print alignment
def printalignment(align):
  for x in align:
    print(fa(*x))

# global
global_align = pairwise2.align.globalxx(seqA, seqB)
printalignment(global_align)
# local
local_align = pairwise2.align.localxx(seqA, seqB)
printalignment(local_align)

# distance (pake hamming / levensht)
```

```python
# plotting
plt.pie([freqA['T'],freqA['A'],freqA['G'],freqA['C']], labels=['T','A','G','C'])
plt.title('freq seqA')
plt.show()

plt.pie([freqB['T'],freqB['A'],freqB['G'],freqB['C']], labels=['T','A','G','C'])
plt.title('freq seqB')
plt.show()
```
